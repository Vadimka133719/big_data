```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from pylab import *
%matplotlib inline
```

## Зависимость объёма кредита от среднего количетсва покупок

```python
purchases = np.random.normal(3.0,1.0,1000)
credit = 25 + (purchases + np.random.normal(0,0.5,1000))*3
```

## Диаграмма рассеяния

```python
scatter(purchases,credit)
```

## Коэффициент корреляции

```python
np.corrcoef(purchases,credit)
```

## Линейная рагрессия

```python
slope, intercept, _, _, _ = stats.linregress(purchases,credit)
```

```python
def predict(x):
    return slope*x+intercept
```

```python
fitLine = predict(purchases)
```

```python
plt.scatter(purchases,credit)
plt.plot(purchases,fitLine,c='r')
plt.show()
```

## Прогноз

```python
purchases_ = 7
```

```python
print('Кредит для кол-ва покупок', purchases_, ':', predict(purchases_))
```
