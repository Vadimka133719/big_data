```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import * 
```

## Зависимость объема кредита от среднего количества покупок

```python
purchases = np.random.normal(3.0,1.0,1000)
credit = 25 + (purchases + np.random.normal(0,0.5,1000)) * 3
```

## Диаграмма рассеяния

```python
scatter(purchases,credit)
```

## Коэффициент корреляции

```python
np.corrcoef(purchases,credit)
```

## Линейная регрессия

```python
slope, intercept, _, _, _ = stats.linregress(purchases,credit)
```

```python
def predict(x):
    return slope*x+intercept
```

```python
fitLine = predict(purchases)
```

```python
plt.scatter(purchases,credit)
plt.plot(purchases,fitLine, c='r')
plt.show()
```

## Прогноз

```python
purchases_ = 7
```

```python
print('Кредит для количества покупок', purchases_, ':', predict(purchases_))
```
