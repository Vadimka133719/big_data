```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-4,4,0.01)
print(x)
```

```python
plt.plot(x,x**2)
```

```python
plt.plot(x,x**2)
plt.plot(x,x**3)
```

```python
values = [12,55,4,32,14]
colors = ['r','g','b','c','gray']
plt.bar(range(0,5),values,color = colors)
```

```python
values = [12,55,4,32,14]
colors = ['r','g','b','c','m']
explode = [0,0,0.2,0,0]
labels = ['one','two','three','four','five']
plt.title('Diagram')
plt.pie(values, colors = colors, labels = labels, explode = explode)
```
